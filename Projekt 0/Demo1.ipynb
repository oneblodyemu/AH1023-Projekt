## Hantera data mha Pandas

```python
# Läs in en fil med .csv data i en tabell
df = pd.read_csv('demo1.csv')

# Namnge kolumnen i tabellen med valfritt namn
df.columns = ['data']

# Visa tabellen
df.head()
```

```python
# Beräkna medel ch median och skriv ut resultatet
print(df['data'].mean())
print(df['data'].median())
```

```python
# Eller i två steg, genom att spara resultatet som en variabel
mean = df['data'].mean()
median = df['data'].median();
print(mean)
print(median)
```

Notera att vi skulle kunna skriva ut en av dem (tex medelvärdet) utan att använda print, genom att bara skriva `mean` på sista raden i cellen. Med `print()` får vi större kontroll över utskriften.

```python
df.describe()
```

```python
# Ibland är det lättare att läsa tabellen om vi byter plats på axlarna
df.describe().transpose()
```

```python
# Snabbkommando för att visa en rad statistiska värden för tabellen
display(df.describe().transpose())

#Standard för percentilerna är 25%, 50%, ch 75%. Sätt egna percentiler såhär:
percs = [.20, .50, .80]
display(df.describe(percentiles=percs).transpose())
```

```python
# Om tabellen hade innehållit flera kolumner och vi bara vill ha statstik på en av dessa, kan man t ex gröpa ut datat i denna kolumnen,
# göra denna till en dataframe och sedan ta describe på detta.
display(df['data'].to_frame().describe())
```

## Plocka ut delar av data

```python
# Sätter "True" eller "False" för varje rad som uppfyller villkoret, här att vara mindre än 45. Detta sparas i nya tabellen radindex_under_45
radindex_under_45 = df['data'] < 45
radindex_under_45.head(5)
```

```python
# Plockar ut raderna med "True" och lägger in dem i en ny tabell
df_below_45 = df[radindex_under_45]

# Visa nya tabellen som nu exkluderar värdet 45 (rad 13)
display(df_below_45)
```

## Visualisera

```python
# Visa en boxplot.
# Data vi använder här tas från tabellen df. "data" motsvarar tabellens kolumnnamn som vi vill använda.
fig = px.box(df, y="data") #df kan t.ex. bytas ut mot df_below_45 som då i stället visar en boxplot på tabellen som exluderar värdet 45

# En figure kan uppdateras och anpassas på olika vis genom fig.update_layout

# Möjliga teman på plottens utseende kan väljas från någon av
# "plotly", "plotly_white", "plotly_dark", "ggplot2", "seaborn", "simple_white", "none"
fig.update_layout(template="plotly_white", title="Box plot")
fig.show()
```

```python
#Violin plot är en liknande variant
fig = px.violin(df, y="data", box=True, template="ggplot2", title="Violin plot") #Övriga inställningar kan också läggas in direkt, som här
fig.show()
```

```python
fig = px.histogram(df, x="data", nbins=5, title="Histogram") #Plotly korrigerar ibland +/- 1 bins utefter vad den själv tycker blir bäst
fig.show()
```

## Scatterplot och regression



```python
# Skapa lite nya data

mu, sigma = 0, 0.1 # mean and standard deviation

# Skapa en array med 20 värden från 0 upp till 10 (9.5 blir då sista värdet),
# med intervallet 0.5
xval = np.arange(0, 10, 0.5)
yval = 1 + 0.35 * xval + np.random.normal(mu, sigma, len(xval))
#y-värdena blir en funktion av x värdena med random gaussian noise (normalfördelat brus)

#Lägg in x-värden och y-värden i varsin kolumn i en tabell
df2 = pd.DataFrame({"xval": xval, "yval": yval})

# Plotta med t ex plotly express
fig = px.scatter(df2, x='xval', y='yval', title="Scatter plot", trendline="ols")

# trendline lägger till least squares linjen

fig.show()
```

```python
# lägger till en konstant för att skatta intercept
df2['const'] = 1

# väljer variabler för regressionen
Y = df2['yval']
X = df2[['const', 'xval']]

model = sm.OLS(Y, X).fit()

print(model.summary())
```
