```python
import geopandas as gpd
import plotly.express as px;
import pandas as pd

dagbef_tid_area_path = r"C:\Users\fabia\AH1023\Projekt\Projekt_2\data2\dagbef_area.xlsx"
bef_tid_lang_path = r"C:\Users\fabia\AH1023\Projekt\Projekt_2\data2\be0101_folkmangdkom2024.xlsx"
bef_tid_kort_path = r"C:\Users\fabia\AH1023\Projekt\Projekt_2\data2\befolkningsfoeraendringar-helar.parquet"
nattbef_tid_area_path = r"C:\Users\fabia\AH1023\Projekt\Projekt_2\data2\nattbef_area.xlsx"
#visualisering av umeås historiska befolkningsutveckling användninga av scatter plot
bef_tid_kort = pd.read_parquet(bef_tid_kort_path)
nattbef_tid_area = pd.read_excel(nattbef_tid_area_path)
bef_tid_lang = pd.read_excel(bef_tid_lang_path)
dagbef_tid_area = pd.read_excel(dagbef_tid_area_path)
# select Excel row 245 (1-based)
bef_tid_lang = bef_tid_lang.iloc[[4,243]]
# Extract years from row 4 and population data from row 243
years = bef_tid_lang.iloc[0, 1:].values
population = bef_tid_lang.iloc[1, 1:].values

# Create scatter plot
fig = px.line(x=years, y=population, labels={'x': 'År', 'y': 'Folkmängd'}, title='Umeås befolkningsutveckling 1950-2024')
fig.show()
```

Analys av visualisering

```python
#visualisering av befolkning med de olika kranskommunerna: Bjurholm, Nordmaling, Robertsfors , Vindeln, Vännäs
bef_tid_region = pd.read_excel(bef_tid_lang_path)

# Välj rätt rader för de 5 kommunerna (utan årtalsraden)
bef_tid_region = bef_tid_region.iloc[[17, 159, 181, 260, 264]]

# År ligger i den första raden av originalfilen (innan vi filtrerade)
bef_tid_lang_all = pd.read_excel(bef_tid_lang_path)
years = bef_tid_lang_all.iloc[4, 1:].values  # Rad 4 innehåller årtalen

municipalities = ['Bjurholm', 'Nordmaling', 'Robertsfors', 'Vindeln', 'Vännäs']
population_data = {}

# Läs befolkning från rätt rader
for idx, municipality in enumerate(municipalities):
    population_data[municipality] = bef_tid_region.iloc[idx, 1:].values

# Create line plot with plotly
fig = px.line(x=years, y=[population_data[m] for m in municipalities], 
              labels={'x': 'År', 'y': 'Folkmängd'}, 
              title='Befolkningsutveckling i omkringliggande kommuner kring Umeå 1950-2024')

# Rename traces with municipality names
for i, municipality in enumerate(municipalities):
    fig.data[i].name = municipality

fig.show()

```
