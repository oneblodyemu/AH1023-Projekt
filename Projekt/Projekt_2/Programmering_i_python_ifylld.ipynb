## Listor
En lista i Python är en inbyggd datatyp som används för att lagra en samling av element




*   Ordnade: Indexerbara (bröjar på 0)

## Logiska operationer: Kontrollera flödet med if-elif-else-satser


*   **if**:   Kontrollerar om ett villkor är sant och kör då koden inuti.
*   **elif**: Testar ett nytt villkor om det första är falskt.
*   **else**: Kör koden här om inga tidigare villkor var sanna.






````python
e = 400  # Variabeln e tilldelas värdet 400

if e == 100:  # Kontrollerar först om e är lika med 100
    print("Ja, e har värdet 100")  # Skriver ut detta meddelande om villkoret är sant
elif e == 200:  # Om det första villkoret är falskt, kontrollerar om e är lika med 200
    print("Ja, e har värdet 200")  # Skriver ut detta meddelande om detta villkor är sant
elif e == 300:  # Kontrollerar om e är lika med 300
    print("Ja, e har värdet 300")  # Skriver ut detta meddelande om detta villkor är sant
else:  # Om ingen av de tidigare villkoren stämmer
    print("e har varken värdet 100, 200 eller 300")  # Skriver ut detta meddelande
````

## Loopar: Upprepa så länge ett villkor är sant med while-satsen
Utför en åtgärd upprepade gånger tills ett specifikt villkor blir falskt.

````python
x = 0
while x < 3: #Loopen skriver ut värdet av x och ökar det med 1 varje gång, tills x når 3 och villkoret x < 3 inte längre är sant.
    print(x)
    x += 1
````

## Funktioner: Definiera återanvändbara kodblock
Funktioner i Python är block av kod som kan definieras för att utföra en specifik uppgift. Genom att kapsla in kod i funktioner kan vi återanvända den flera gånger utan att behöva skriva om den. Funktioner kan ta emot indata (argument) och kan returnera värden.

````python
def addera(x, y):
    z = x + y
    return z

print(addera(2, 3))
print(addera(1000, 10000))
````

## NumPy

Ett av de vanligaste biblioteken i Python är NumPy, som ger snabb och effektiv hantering av vektorer och matriser.

Bibliotek importeras med "import":

Vår nyskapta array, **arr**, är ett så kallat **objekt**. Från detta objekt kan vi använda både **attribut** och funktioner (eller **metoder**, som funktionerna också kallas när de är definierade för ett objekt).

## Attribut

*   Definition: Attribut är variabler kopplade till ett objekt (eller en klass) som lagrar data eller tillstånd (egenskaper) hos det objektet.
*   Exempel: För en NumPy-array finns attribut som t ex shape eller ndim.
*   Åtkomst: Attribut hämtas med punktnotation, följt av attributets namn. Exempel:

````python
print(arr.shape)  # Skriver ut (2, 3) - antalet rader och kolumner
print(arr.ndim)   # Skriver ut antalet dimensioner, t.ex. 2
````

<br><br>
dir() är en inbyggd funktion i Python som används för att lista attributen och metoderna för ett objekt.

## Metoder (funktioner)
*   Definition: Funktioner är block av kod som utför specifika operationer eller beräkningar relaterade till objektets data.
*   Exempel: Med en NumPy-array kan du använda metoder som sum(), mean(), reshape(), och många fler för att utföra olika operationer.
*   Åtkomst: Metoder anropas med punktnotation följt av metodens namn och parenteser. Exempel:

## Addition av Arrays jämfört med Listor
Addition av arrays i NumPy fungerar annorlunda än hur listor läggs ihop i Python. Här är en jämförelse:
### Python-listor
När vi lägger ihop två listor med hjälp av +, slås de samman till en ny lista. Det betyder att elementen i de två listorna sammanfogas. Här är ett exempel:

````python
l1 = [0, 0, 0]
l2 = [1, 1, 1]
l3 = l1 + l2
print(l3)
````

<br><br>

### NumPy-arrayer
I kontrast till detta, när du lägger ihop två NumPy-arrayer med +, utförs en elementvis addition.

````python
arr1 = np.array([0, 0, 0])  # Första arrayen
arr2 = np.array([1, 1, 1])  # Andra arrayen
arr3 = arr1 + arr2          # Elementvis addition av arrayerna
print(arr3)                 # Skriver ut [1 1 1]
````

Alternativt kan arrayerna skapas med hjälp av np.zeros() och np.ones()

````python
arr1 = np.zeros(3)  # Skapar en array med tre nollor
arr2 = np.ones(3)   # Skapar en array med tre ettor
arr3 = arr1 + arr2  # Elementvis addition av arrayerna
print(arr3)         # Skriver ut [1. 1. 1.]
````

Följande operationer påverkar alla element i en array

````python
arr4 = arr3 + 1
print(arr4)
arr5 = arr3 * 10
print(arr5)
````

## Pandas

Ett annat omåttligt populärt bibliotek i python är tabellhanteraren Pandas, designat för att underlätta datahantering och analys.

#### Vad är Pandas?

Pandas introducerar **DataFrame** - en tvådimensionell tabellstruktur som är storleksändringsbar och som liknar en Excel-tabell.

Vi inleder med att skapa en egen enkel tabell som vi kallar df, förkortning av dataframe (tabell), men kunde såklart döpas till något annat namn.



````python
import pandas as pd

# Skapar listor med värden
val1 = [1, 2, 3, 4, 5, 6]
val2 = [11, 12, 13, 14, 15, 16]

# Skapar en ordlista med kolumnnamn
data = {"kolumn 1": val1, "kolumn 2": val2}

# Skapar en DataFrame baserat på ordlistan data ovan
df = pd.DataFrame(data)

print(df)    # print() fungerar, men kan vara svårläst
display(df)  # Mer lättläst blir det med display()
df.head()    # Metoden .head() skriver ut de fem första raderna av tabellen (.tail() visar de fem sista)
````

## Plotly

Det finns en uppsjö av olika bibliotek för plottning i Python. Ett av de mest populära är **Plotly**, som erbjuder interaktiva grafer och visualiseringar.
Vi tar vår tidigare tabell df2 med data från demo.csv, och plottar det i ett histogram.

````python
import plotly.express as px
fig = px.histogram(df2, x="data", nbins=8, title="Histogram") #Plotly korrigerar ibland +/- 1 bins utefter vad den själv tycker blir bäst
fig.show()
````

Vi skapar lite syntetisk data för plottning och analys med trendlinje. Syntetisk data är artificiellt genererad snarare än insamlad från verkliga händelser, och är ett effektivt sätt att experimentera och analysera olika metoder.

````python
xval = np.arange(0, 10, 0.5) # Vår oberoende variabel

mu, sigma = 0, 0.5 # Medelvärde och standardavvikelse
yval = 1 + 0.35 * xval + np.random.normal(mu, sigma, len(xval))

#y-värdena blir en funktion av x värdena med random gaussian noise (normalfördelat brus)
````

Sedan lägger vi in våra x-värden och y-värden i varsin kolumn i en tabell, redo att skicka med i en scatterplot, som dessutom kan rita ut en trendlinje baserat på Ordinary Least Squares (OLS).

````python
#Lägg in x-värden och y-värden i varsin kolumn i en tabell
df2 = pd.DataFrame({"xval": xval, "yval": yval})

# Plotta med t ex plotly express
fig = px.scatter(df2, x='xval', y='yval', title="Scatter plot", trendline="ols") # trendline lägger till least squares linjen

#OLS anpassar en linjär modell till data, genom att minimera summan av kvadraterna av skillnaderna mellan de observerade värdena och de värden som förutsägs av modellen.
#y = a + b*x

#fig.update_yaxes(range=[0, 100])
#fig.update_layout(xaxis_title="En stads befolkningstäthet", yaxis_title="Andel resor med kollektivtrafik")

fig.show()
````
